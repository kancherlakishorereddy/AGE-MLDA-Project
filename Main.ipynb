```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('age_gender.csv');
```

```python
X = pd.DataFrame(df['pixels'].str.split(' ', 48*48-1, expand=True)).astype(float)
```

```python
plt.figure(figsize=(16,16))
for idx, i in enumerate([200, 2000, 12000, 23000]):
    plt.subplot(5,5,idx+1)
    plt.grid(False)
    plt.imshow(X.iloc[i].values.reshape(48,48))
    plt.xlabel(
        "Age:"+str(df['age'].iloc[i])+
        "  Ethnicity:"+str(df['ethnicity'].iloc[i])+
        "  Gender:"+ str(df['gender'].iloc[i])
    )
plt.show()
```

```python
age_dist = df['age'].value_counts()
gender_dist = df['gender'].value_counts()
eth_dist = df['ethnicity'].value_counts()
```

```python
# Smoothed Histogram Age distribuition
fig, ax = plt.subplots()
pd.Series(df['age']).plot.kde(ax = ax)
pd.Series(df['age']).plot.hist(density=True, ax=ax, bins = 115, figsize=(10,7))
plt.xlabel('Attr 3')
plt.show()
```

```python
# Bar Graph Ethnicity Distribuition
plt.bar(eth_dist.index, eth_dist.values)
plt.show()
```
